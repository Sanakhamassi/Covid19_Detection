```python
import os
import shutil
from sklearn.model_selection import train_test_split
```

# Split dataset

```python
def split_dataset:
    # Set the path to your dataset directory
    dataset_path = "./Dataset"
    #Defining paths
    train_dir = "covidDataset/Train"
    test_dir = "covidDataset/Test"
    # Create train and test directories if they don't exist
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)
    # Get the list of subdirectories (covid, normal, pando)
    subdirectories = [d for d in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, d))]
    # Iterate through each subdirectory
    for subdir in subdirectories:
        # Get the list of files in the current subdirectory
        files = os.listdir(os.path.join(dataset_path, subdir))
        # Split the files into train and test sets
        train_files, test_files = train_test_split(files, test_size=0.2, random_state=42)
        # Create subdirectories in train and test directories
        os.makedirs(os.path.join(train_dir, subdir), exist_ok=True)
        os.makedirs(os.path.join(test_dir, subdir), exist_ok=True)
        # Copy files to train directory
        for file in train_files:
            src = os.path.join(dataset_path, subdir, file)
            dest = os.path.join(train_dir, subdir, file)
            shutil.copy(src, dest)
        # Copy files to test directory
        for file in test_files:
            src = os.path.join(dataset_path, subdir, file)
            dest = os.path.join(test_dir, subdir, file)
            shutil.copy(src, dest)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense,BatchNormalization,Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Définition du générateur de données pour l'entraînement avec prétraitement
train_datagen = ImageDataGenerator(
    rescale=1./255,  # normalisation des valeurs de pixel
    rotation_range=20,  # exemple d'augmentation de données (rotation)
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)

# Définition du générateur de données pour le test sans augmentation
test_datagen = ImageDataGenerator(rescale=1./255)

# Chargement des images à partir du répertoire et application du prétraitement
train_generator = train_datagen.flow_from_directory(
    'covidDataset/Train',
    target_size=(256, 256),  # redimensionnement des images
    batch_size=32,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    'covidDataset/Test',
    target_size=(256, 256),
    batch_size=32,
    class_mode='categorical'
)

# Construction du modèle CNN
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2))) 
# Ajoutez plus de couches convolutives, de normalisation et de pooling si nécessaire...
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))  # Example dropout layer
model.add(Dense(3, activation='softmax'))

# Compilation du modèle
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Entraînement du modèle
model.fit(train_generator, validation_data=test_generator, epochs=10)  # Modifier le nombre d'époques selon les besoins
```

```python
model.summary()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
# Prédiction des classes pour les images de test
predictions = model.predict(test_generator)

# Récupération des classes prédites (indices des classes ayant la plus haute probabilité)
predicted_classes = predictions.argmax(axis=1)

# Récupération des vraies classes des images de test
true_classes = test_generator.classes

# Création d'un dictionnaire pour mapper les indices de classe aux noms de classe
class_mapping = {0: 'COVID', 1: 'NORMAL', 2: 'PNEUMONIA'}

# Mapper les classes prédites et les vraies classes aux noms de classe correspondants
predicted_classes_names = [class_mapping[class_idx] for class_idx in predicted_classes]
true_classes_names = [class_mapping[class_idx] for class_idx in true_classes]

# Affichage des noms de classes prédites et des vrais noms de classes (pour les 10 premières)
print("Classes prédites :", predicted_classes_names[:10])
print("Vraies classes :", true_classes_names[:10])

# Assuming you have a function to load images from file paths
def load_image(img_path):
    img = image.load_img(img_path, target_size=(256, 256))  # Adjust target_size as needed
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    return img_array

# Assuming you have the file paths for the test images
test_image_paths = test_generator.filepaths

# Number of images to display
num_images_to_display = 10

# Loop through the first few test images
for i in range(num_images_to_display):
    # Load the image
    img_path = test_image_paths[i]
    img = load_image(img_path)
    
    # Display the image
    plt.imshow(np.uint8(img[0]))
    
    # Get the predicted and true classes
    predicted_class = predicted_classes[i]
    true_class = true_classes[i]
    
    # Map the class indices to class names
    predicted_class_name = class_mapping[predicted_class]
    true_class_name = class_mapping[true_class]
    
    # Display the predicted and true classes
    plt.title(f"Predicted: {predicted_class_name}\nTrue: {true_class_name}")
    
    # Show the plot
    plt.show()
```

```python
# Évaluation de l'exactitude du modèle
accuracy = (predicted_classes == true_classes).sum() / len(true_classes)
print("Exactitude du modèle :", accuracy)
# Importation de confusion_matrix pour créer une matrice de confusion
from sklearn.metrics import confusion_matrix
# Création de la matrice de confusion
conf_matrix = confusion_matrix(true_classes, predicted_classes)
print("Matrice de confusion :\n", conf_matrix)
```
